```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat

# Load the .mat file
mat_data = loadmat('./dataGRFshorttspan.mat')  # replace with your actual .mat file path

# Set seed and sample random indices
np.random.seed(0)
indices = np.random.randint(0, 1200, size=3)

# High-quality figure for publication
plt.figure(figsize=(6, 3), dpi=300)

for idx in indices:
    aa = mat_data['output'][idx, :]
    plt.plot(np.linspace(0, 1, 8192, endpoint=False), aa, linewidth=1)

plt.xlabel("x", fontsize=10)
plt.ylabel(r"$u_0(x)$", fontsize=10, rotation=0, labelpad=10)
plt.title("Examples of initial conditions", fontsize=11)
plt.grid(True, linestyle="--", linewidth=0.5)
plt.tight_layout()
plt.show()
```
